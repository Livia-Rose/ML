```python
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, roc_auc_score
from sklearn.preprocessing import LabelEncoder
import numpy as np

# Load the dataset from Colab's file system
# Note: You need to upload the full "UPI apps transaction data in 2021.csv" to /content/
file_path = '/content/sample_data/UPI apps transaction data in 2021.csv'  # Adjust path if needed
data = pd.read_csv(file_path)

# Preprocess the data
# Define numerical and categorical features
numerical_features = ['Volume (Mn) By Costumers', 'Value (Cr) by Costumers', 'Volume (Mn)', 'Value (Cr)', 'Month', 'Year']
categorical_features = ['UPI Banks']

# Handle missing values
data[numerical_features] = data[numerical_features].fillna(data[numerical_features].median(numeric_only=True))

# Handle categorical features with LabelEncoder
label_encoders = {}
for col in categorical_features:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col].astype(str))
    label_encoders[col] = le

# Select features for Isolation Forest
X = data[numerical_features + categorical_features]

# Split data into train and test sets (for evaluation; Isolation Forest is unsupervised)
# Here, we'll use a proxy for anomalies, e.g., low volume transactions (< median volume)
y_proxy = (data['Volume (Mn)'] < data['Volume (Mn)'].median()).astype(int)  # 1 for low volume (potential anomaly), 0 otherwise
X_train, X_test, y_train_proxy, y_test_proxy = train_test_split(X, y_proxy, test_size=0.2, random_state=42)

# Initialize Isolation Forest model
model = IsolationForest(
    n_estimators=100,        # Number of trees
    contamination='auto',    # Automatically estimate outlier proportion
    max_samples='auto',      # Use all samples or a fraction
    random_state=42          # For reproducibility
)

# Fit the model on the training data
model.fit(X_train)

# Predict anomalies on the test set (-1 for anomaly, 1 for normal)
anomalies = model.predict(X_test)

# Map predictions to binary labels: 1 for anomaly, 0 for normal
y_pred = [1 if x == -1 else 0 for x in anomalies]

# Evaluate the model using proxy labels (low volume as anomalies)
print(f"ROC AUC Score (using low volume as proxy for anomalies): {roc_auc_score(y_test_proxy, y_pred):.4f}")
print("\nClassification Report:")
print(classification_report(y_test_proxy, y_pred))

# Save the model for future use
import joblib
joblib.dump(model, '/content/isolation_forest_upi_model.pkl')

# To detect anomalies in new data:
# new_data = pd.read_csv('/path/to/new_data.csv')  # Load new data
# new_data[numerical_features] = new_data[numerical_features].fillna(new_data[numerical_features].median(numeric_only=True))
# new_data['UPI Banks'] = label_encoders['UPI Banks'].transform(new_data['UPI Banks'].astype(str))
# anomalies_new = model.predict(new_data[numerical_features + categorical_features])
# pred_new = [1 if x == -1 else 0 for x in anomalies_new]
# print("New data predictions (1=anomaly, 0=normal):", pred_new)
```

```python
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import roc_auc_score, classification_report
import numpy as np
import pickle

# Load the dataset
file_path = '/content/sample_data/PhoneRecharge_v2.csv'  # Adjust path if needed
data = pd.read_csv(file_path)

# Clean column names
data.columns = data.columns.str.strip()

# Define numerical and categorical features
numerical_features = ['Recharge_Amount']
categorical_features = ['User_ID', 'Name', 'MSISDN', 'Recharge_Type', 'Payment_Mode', 'Region_Language']
date_features = ['Recharge_Date']

# Convert date feature
data['Recharge_Date'] = pd.to_datetime(data['Recharge_Date'], errors='coerce')

# Feature engineering: Extract date components
data['Recharge_Year'] = data['Recharge_Date'].dt.year
data['Recharge_Month'] = data['Recharge_Date'].dt.month
data['Recharge_Day'] = data['Recharge_Date'].dt.day
numerical_features.extend(['Recharge_Year', 'Recharge_Month', 'Recharge_Day'])

# Feature engineering: Add frequency of recharges per user
data['Recharge_Frequency'] = data.groupby('User_ID')['User_ID'].transform('count')
numerical_features.append('Recharge_Frequency')

# Handle missing values
data[numerical_features] = data[numerical_features].fillna(data[numerical_features].median(numeric_only=True))

# Normalize numerical features
scaler = StandardScaler()
data[numerical_features] = scaler.fit_transform(data[numerical_features])

# Encode categorical features
label_encoders = {}
for col in categorical_features:
    le = LabelEncoder()
    data[f'{col}_Encoded'] = le.fit_transform(data[col].astype(str))
    label_encoders[col] = le
    numerical_features.append(f'{col}_Encoded')

# Prepare features for Isolation Forest
X = data[numerical_features]

# Train Isolation Forest
model = IsolationForest(
    n_estimators=200,  # Increased for better robustness
    contamination=0.1,  # Set to 10% anomalies, adjustable
    random_state=42
)
model.fit(X)

# Predict anomalies (-1 for anomaly, 1 for normal)
anomalies = model.predict(X)
data['Anomaly'] = [1 if x == -1 else 0 for x in anomalies]

# Create a proxy label for evaluation (high Recharge_Amount as anomalies)
y_proxy = (data['Recharge_Amount'] > data['Recharge_Amount'].quantile(0.95)).astype(int)
y_pred = data['Anomaly']

# Evaluate using proxy labels
print(f"ROC AUC Score (using high Recharge_Amount as proxy for anomalies): {roc_auc_score(y_proxy, y_pred):.4f}")
print("\nClassification Report:")
print(classification_report(y_proxy, y_pred))

# Export the models and results
# 1. Save the Isolation Forest model
with open('/content/isolation_forest_model.pkl', 'wb') as f:
    pickle.dump(model, f)

# 2. Save the StandardScaler
with open('/content/recharge_scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)

# 3. Save the LabelEncoders
for col, le in label_encoders.items():
    with open(f'/content/recharge_label_encoder_{col}.pkl', 'wb') as f:
        pickle.dump(le, f)

# 4. Save the dataset with anomaly labels
# Only inverse_transform categorical columns
for col in categorical_features:
    data[col] = label_encoders[col].inverse_transform(data[f'{col}_Encoded'])
data.to_csv('/content/PhoneRecharge_with_anomalies.csv', index=False)

# Print anomalies
anomalies_df = data[data['Anomaly'] == 1]
print("\nDetected anomalies:")
print(anomalies_df[['User_ID', 'Name', 'MSISDN', 'Recharge_Amount', 'Recharge_Type', 'Payment_Mode', 'Anomaly']])
```
